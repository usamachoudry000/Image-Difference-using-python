```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./dataset/img4.JPG')
img.shape
```

```python
resized_img=cv2.resize(img,(500,500),2)
resized_img.shape
```

```python
plt.title('BGR image')
plt.imshow(resized_img)
```

```python
import random
m=1
n=random.randint(1,5)

height=resized_img.shape[0]
width=resized_img.shape[1]
crop = np.ndarray((5,),dtype=object)
crop[0]=resized_img[0:height//2,0:width//2]
crop[1]=resized_img[0:height//2,width//2:width]
crop[2]=resized_img[height//2:height,0:width//2]
crop[3]=resized_img[height//2:height,width//2:width]
crop[4]=resized_img[0:height//2,width//2:width]    
for i in range(m*n):
    axis=plt.subplot(m,n,i+1)
    plt.title('patches'+str(i))
    plt.axis('off')
    plt.imshow(crop[i])
    plt.imsave('./gallery/pic{:>01}.jpg'.format(i),crop[i])
```

```python
import random
n=1
m=random.randint(1,5)

height=resized_img.shape[0]
width=resized_img.shape[1]
crop = np.ndarray((5,),dtype=object)
crop[0]=resized_img[0:height//2,0:width//2]
crop[1]=resized_img[0:height//2,width//2:width]
crop[2]=resized_img[height//2:height,0:width//2]
crop[3]=resized_img[height//2:height,width//2:width]
crop[4]=resized_img[0:height//2,width//2:width]    
for i in range(m*n):
    axis=plt.subplot(n,m,i+1)
    plt.title('patches'+str(i))
    plt.axis('off')
    plt.imshow(crop[i])
    plt.imsave('./query/pic{:>01}.jpg'.format(i),crop[i])
```

```python
gallery = []
query=[]
for i in range(n):
    gimg=cv2.imread('./gallery/pic{:>01}.jpg'.format(i))
    gallery.append(gimg)
    #plt.imshow(gallery)
#print(gimg)


for i in range(m):
    qimg=cv2.imread('./query/pic{:>01}.jpg'.format(i))
    query.append(qimg)
#print(query)
```

```python
#Difference
diff=[]
for i in range(m):
    for j in range(n):
        d1=query[i]
        d2=gallery[j]
        diff.append(cv2.subtract(d1,d2)**2)
        #plt.imsave('./difference/pic{:>01}.jpg'.format(j),diff[j])
        
print(diff)
        
```

```python
for i in range(m*n):
    #final=cv2.imread('./difference/pic{:>01}.jpg'.format(i))
    axis=plt.subplot(m,n,i+1)
    plt.title('Difference'+str(i))
    plt.axis('off')
    plt.imshow(diff[i])
```
